```python
#Create a data frame from list using dictionary
dicc = {'Pid': pid,'Element': elements}
print(dicc)


```

```python
dfsorted = df_drop.sort_index(axis=0,ascending=True)
#print(dfsorted)
```

```python
print(dfsorted.to_string())
```

```python
#Add elements as columns with value NaT
dfsorted[elements]  = "." #pd.NaT
print(dfsorted.to_string())
```
